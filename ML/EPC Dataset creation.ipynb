# Image Extraction

```python
frame_skip_factor = 4

for label in labels:
    # path of each video and image folder
    label_video_path = os.path.join(video_path, label)
    label_img_path = os.path.join(img_path, label)
    
    for pose in os.listdir(label_video_path):
        file_path = os.path.join(label_video_path, pose)
        save_path = os.path.join(label_img_path,pose)
        for filename in os.listdir(file_path):
            # image name
            img_root_name = str(filename).replace(".mp4", "")
            # path of the video file
            filenm_path = os.path.join(file_path, filename)
            webcam = cv2.VideoCapture(filenm_path)
            currentframe = 0
            while True:
                # reading from frame
                ret, frame = webcam.read()

                if ret:
                    # if video is still left continue creating images
                    name = img_root_name + str(currentframe) + '.jpg'
                    print('Creating...' + name)

                    # writing the extracted images
                    cv2.imwrite(os.path.join(save_path, name), frame)
                    
                    # Skip frames based on the frame_skip_factor
                    for _ in range(frame_skip_factor - 1):
                        _ = webcam.read()

                    # show how many frames are created
                    currentframe += 1
                else:
                    break

            webcam.release()

cv2.destroyAllWindows()
```
