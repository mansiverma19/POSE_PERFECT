```python
 
def calculate_angle(A, B, C):
    # Convert points to numpy arrays
    A = np.array(A)
    B = np.array(B)
    C = np.array(C)
   # Calculate the distances between points
    AB = np.linalg.norm(B - A)
    BC = np.linalg.norm(C - B)
    AC = np.linalg.norm(C - A)
    
    # Use the law of cosines to calculate the angle at point B
    radians = np.arccos((AB**2 + BC**2 - AC**2) / (2 * AB * BC))
    angle_degrees = np.degrees(radians)
    return angle_degrees



def get_imp_keypnt(Category, Pose, final_jnt_lst):
    dct = {}
    final_dct = {}
    if Category == 'Exercise':
        if Pose in exr_df['EXERCISE NAME'].values:
            # Filter the DataFrame based on the target_pose
            result_df = exr_df[exr_df['EXERCISE NAME'] == Pose]
            for jnt in list(exr_df.columns)[5:]:
                dct[jnt] = result_df[jnt].values[0]
    elif Category == 'Yoga':
        if Pose in yoga_df['YOGA NAME'].values:
            result_df = yoga_df[yoga_df['YOGA NAME'] == Pose]
            for jnt in list(yoga_df.columns)[5:]:
                dct[jnt] = result_df[jnt].values[0]
    else:
        print(f"Pose '{Pose}' not found in the DataFrame.")
        return None
    main_key_lst = list(dct.keys())
    for nm in main_key_lst:
        for i in final_jnt_lst:
            if nm in i:
                final_dct[i] = dct[nm]
    return final_dct            
        
```

```python
import mediapipe as mp
import cv2

# Initialize MediaPipe Pose model
pose = mp.solutions.pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Process each image
for i in range(12, 23):
    img_path = str(i) + '.jpg'
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = pose.process(img_rgb)
    print(img.shape)
    
    # adjusting larger images 
    ht = img.shape[0]
    wdt = img.shape[1]
    if ht > wdt:
        ratio = ht / wdt
        desired_width = 400
        desired_height = int(ratio*desired_width)
    else:
        ratio = wdt / ht
        desired_height = 400
        desired_width = int(ratio*desired_height)

    # Resize the image
    img = cv2.resize(img, (desired_width, desired_height))

    # Check if pose landmarks are detected
    if result.pose_landmarks:
        # Extract pose landmarks
        landmarks = result.pose_landmarks.landmark
        
        # Get the coordinates of the shoulders, hips, and nose
        left_shoulder = landmarks[mp.solutions.pose.PoseLandmark.LEFT_SHOULDER.value]
        right_shoulder = landmarks[mp.solutions.pose.PoseLandmark.RIGHT_SHOULDER.value]
        left_hip = landmarks[mp.solutions.pose.PoseLandmark.LEFT_HIP.value]
        right_hip = landmarks[mp.solutions.pose.PoseLandmark.RIGHT_HIP.value]
        nose = landmarks[mp.solutions.pose.PoseLandmark.NOSE.value]
        
        # Calculate the horizontal midpoint of the shoulders and hips
        shoulder_midpoint = ((left_shoulder.x + right_shoulder.x) / 2, (left_shoulder.y + right_shoulder.y) / 2)
        hip_midpoint = ((left_hip.x + right_hip.x) / 2, (left_hip.y + right_hip.y) / 2)
        
        # Draw circles at the calculated points
        img_with_points = img.copy()
        radius = 5
        color = (0, 255, 0)  # Green color
        
        cv2.circle(img_with_points, (int(shoulder_midpoint[0] * img.shape[1]), int(shoulder_midpoint[1] * img.shape[0])), radius, color, -1)
        cv2.circle(img_with_points, (int(hip_midpoint[0] * img.shape[1]), int(hip_midpoint[1] * img.shape[0])), radius, color, -1)
        cv2.circle(img_with_points, (int(nose.x * img.shape[1]), int(nose.y * img.shape[0])), radius, color, -1)
        
        # Display the image with plotted points
        cv2.imshow('Image with Points', img_with_points)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("Pose landmarks not detected")

# Release resources
pose.close()
```
