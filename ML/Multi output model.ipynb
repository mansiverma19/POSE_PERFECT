```python
num_joints = 13
max_wrong_joints = 5
joint_names = list(df.columns[3:16])
print(joint_names)

# Define function to generate wrong joints based on "Is_Wrong" column
def generate_wrong_joints(row):
    if row['Correct'] == 0:
        num_wrong_joints = np.random.randint(1, max_wrong_joints + 1)  # Random number of wrong joints
        val = np.random.choice(joint_names, size=num_wrong_joints, replace=False)

        # for mapped 
        lst = []
        for i in val:
            if row[i] != -1:
                lst.append(i)
        return ','.join(lst)
    else:
        return ''

# Apply the function to fill the "Wrong_Joints" column
df['Wrong_Joints'] = df.apply(generate_wrong_joints, axis=1)

df.head(15)
```

```python
from sklearn.svm import SVC
from sklearn import set_config
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Splitting the dataset into features and labels
X = df.drop(columns=['Correct','WRONG','Wrong_Joints','Category'])  # Features

y_binary = df['Correct']  # Binary classification target
y_multilabel = df['Wrong_Joints']  # Multi-label classification target
y_binary = y_binary.values

# Convert y_multilabel_train to list of lists
y_multilabel_train_list = [labels.split(',') if labels else [] for labels in y_multilabel]

# Initialize an empty list to store the converted rows
y_multilabel_encoded = []
for labels in y_multilabel_train_list:
    # Initialize a list to store the presence or absence of labels
    row_encoding = []
    # Iterate over each label in all_labels
    for label in list(df.columns[3:16]):
        # Check if the label is present in the current row
        if label in labels:
            row_encoding.append(1)
        else:
            row_encoding.append(0)
    # Append the encoded row to y_multilabel_encoded
    y_multilabel_encoded.append(row_encoding)

# Splitting data into train and test sets
X_train, X_test, y_binary_train, y_binary_test, y_multilabel_train, y_multilabel_test = train_test_split(
    X, y_binary, y_multilabel_encoded, test_size=0.2, random_state=42,shuffle=True
)
y_binary_train[:2], y_multilabel_train[:2]
print(len(y_multilabel_encoded))

```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np

class CustomMultiOutputClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, base_estimator):
        self.base_estimator = base_estimator

    def fit(self, X, y):
        # Extract binary and multi-label targets
        y_binary = y[:, 0].astype(int)
        y_multilabel = y[:, 1:]

        # Fit the binary classifier
        self.base_estimator.fit(X, y_binary)
        
        # Use the binary classifier's predictions as features for the multi-label classifier
        binary_predictions = self.base_estimator.predict(X)
        X_multilabel = np.column_stack((X, binary_predictions))
        
        # Fit the multi-label classifier for each column independently
        self.multioutput_classifiers_ = []
        for i in range(y_multilabel.shape[1]):
            clf = self.base_estimator.__class__()
            clf.fit(X_multilabel, y_multilabel[:, i].astype(int))
            self.multioutput_classifiers_.append(clf)

        # Store the fitted binary classifier
        self.binary_classifier_ = self.base_estimator

        return self

    def predict(self, X):
        # Make binary predictions
        binary_predictions = self.binary_classifier_.predict(X)
        X_multilabel = np.column_stack((X, binary_predictions))
        # Make multi-label predictions for each column independently
        multioutput_predictions = np.zeros((X.shape[0], len(self.multioutput_classifiers_)), dtype=int)
        for i, clf in enumerate(self.multioutput_classifiers_):
            multioutput_predictions[:, i] = clf.predict(X_multilabel)

        # Concatenate binary and multi-label predictions
        predictions = np.column_stack((binary_predictions, multioutput_predictions))

        return predictions
```

```python
from sklearn.feature_extraction import FeatureHasher
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

functionTransformer = FunctionTransformer(lambda x: np.asarray(x.toarray()), accept_sparse=True)
hasher = FeatureHasher(n_features=6, input_type='string')
onehtEncoder = OneHotEncoder()

pipeline_1 = Pipeline([('labelEncoder', onehtEncoder)])  # Define a pipeline for label encoding
ct = ColumnTransformer(
    [('feature hasher', hasher, 0),                   # 0 indexx for 'Pose'    
     ('label encoder', onehtEncoder, [1])],           # 1 index for 'Category'
)

# Training Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=10,max_features="sqrt", random_state=100)
multioutput_classifier_rf = CustomMultiOutputClassifier(rf_classifier)

# Training SVM Classifier
svm_classifier = SVC(kernel='linear')
multioutput_classifier_svm = CustomMultiOutputClassifier(svm_classifier)

# Training KNN
KNN_classifier = KNeighborsClassifier(n_neighbors=3)
# Create the custom multi-output classifier
multioutput_classifier_knn = CustomMultiOutputClassifier(KNN_classifier)

rf_pipe = Pipeline([('column transformer',ct),('func transformer',functionTransformer),('random forest',multioutput_classifier_rf)])
svm_pipe = Pipeline([('column transformer',ct),('func transformer',functionTransformer),('standardscaler',StandardScaler()),('svm',multioutput_classifier_svm)])
knn_pipe = Pipeline([('column transformer',ct),('func transformer',functionTransformer),('KNN',multioutput_classifier_knn)])
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input

# Sample data generation
# Suppose you have a dataset where each row represents joint angles, and columns represent joint names
# Assuming you have some features as well

# Sample dataset
num_samples = 1000
num_features = 10
num_joints = 5
joint_names = [f'Joint_{i}' for i in range(num_joints)]
data = np.random.rand(num_samples, num_features + num_joints)  # Joint angles appended to features
joint_indices = np.random.choice(num_joints, size=(num_samples, 2), replace=False)  # Generating random wrong joint indices
wrong_joints = [joint_names[idx] for idx in joint_indices]

# Create a DataFrame
columns = [f'Feature_{i}' for i in range(num_features)] + joint_names
df = pd.DataFrame(data, columns=columns)
df['Wrong_Joints'] = wrong_joints
df['Is_Wrong'] = 0
df.loc[df['Wrong_Joints'] != '', 'Is_Wrong'] = 1  # Marking rows where wrong joints are present

# Splitting the dataset into features and labels
X = df.drop(columns=['Is_Wrong', 'Wrong_Joints'])  # Features
y_binary = df['Is_Wrong']  # Binary classification target
y_multilabel = df['Wrong_Joints']  # Multi-label classification target

# Multi-label binarization
mlb = MultiLabelBinarizer()
y_multilabel_encoded = mlb.fit_transform(y_multilabel.str.split(','))

# Splitting data into train and test sets
X_train, X_test, y_binary_train, y_binary_test, y_multilabel_train, y_multilabel_test = train_test_split(
    X, y_binary, y_multilabel_encoded, test_size=0.2, random_state=42
)

# Model architecture
inputs = Input(shape=(num_features,))
dense1 = Dense(64, activation='relu')(inputs)
dense2 = Dense(64, activation='relu')(dense1)

# Binary classification output
output_binary = Dense(1, activation='sigmoid', name='binary_output')(dense2)

# Multi-label classification output
output_multilabel = Dense(len(mlb.classes_), activation='sigmoid', name='multilabel_output')(dense2)

# Combine inputs and outputs into a model
model = Model(inputs=inputs, outputs=[output_binary, output_multilabel])

# Compile the model
model.compile(optimizer='adam', loss={'binary_output': 'binary_crossentropy', 'multilabel_output': 'binary_crossentropy'})

# Train the model
model.fit(X_train, {'binary_output': y_binary_train, 'multilabel_output': y_multilabel_train}, epochs=10, batch_size=32)

# Evaluate the model
binary_loss, multilabel_loss = model.evaluate(X_test, {'binary_output': y_binary_test, 'multilabel_output': y_multilabel_test})
print("Binary Classification Loss:", binary_loss)
print("Multi-label Classification Loss:", multilabel_loss)

# Predictions
binary_predictions, multilabel_predictions = model.predict(X_test)

# Decoding multi-label predictions
decoded_predictions = mlb.inverse_transform((multilabel_predictions > 0.5).astype(int))
print("Decoded Multi-label Predictions:", decoded_predictions)
```
